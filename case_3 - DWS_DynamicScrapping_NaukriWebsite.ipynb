# DynamicScrapping_NaukriWebsite

```python
import pandas as pd
import bs4
import selenium
import requests

from selenium import webdriver
```

```python
browser = webdriver.Chrome("chromedriver")
browser.get('https://www.naukri.com/data-science-jobs')
```

```python
# For class,use .
# For others, use #

jobs = browser.find_elements_by_css_selector('div.row[type="tuple"]')
len(jobs)
```

```python
def get_info(job,selector,attribute):
    try:
        tag = job.find_element_by_css_selector(selector)
        tag = tag.get_attribute(attribute)
    except:
        tag = None
    return tag

df_jobs = pd.DataFrame(columns = ['title','company','salary',
                                  'date','location','experience',
                                  'description','URL'
                                 ]
                      )

for job in jobs:
    job_title = get_info(job,'li.desig','title')
    company = get_info(job,'span.org','innerHTML')
    salary = get_info(job,'span.salary','innerHTML')
    date = get_info(job,'span.date','innerHTML')
    try:
        parent_location = job.find_element_by_css_selector('span.loc')
        location = get_info(parent_location,'span','innerHTML')
    except:
        location = None
        
    experience = get_info(job,'span.exp','innerHTML')
    skills = get_info(job,'span.skill','innerHTML')
    description = get_info(job,'span.desc','innerHTML')
    url = get_info(job,'a.content','href')
    
    curr_job = {'title' : job_title,
               'company' : company,
               'salary' : salary,
               'date' : date,
               'location' : location,
               'experience' : experience,
               'description' : description,
               'URL' : url}
    df_jobs = df_jobs.append(curr_job,ignore_index=True)
df_jobs  
```

```python
for url in df_jobs['URL'].head(3):
    browser.get(url)
    full_desc = browser.find_element_by_css_selector('ul.listing.mt10.wb')
    full_desc = full_desc.text
    print(full_desc)
```
