```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.gsmarena.com/apple_iphone_6-reviews-6378.php')
```

```python
comments = browser.find_elements_by_css_selector('div.user-thread')
print(len(comments))
```

```python
df_products = pd.DataFrame(columns = ['user_name','review_date','review_text','upvotes','rating_value'])
for comment in comments:
    try:
        user_name = comment.find_element_by_css_selector('li.uname2')
        user_name = user_name.get_attribute('innerHTML')
    except:
        user_name = comment.find_element_by_css_selector('li.uname')
        user_name = user_name.find_element_by_css_selector('b')
        user_name = user_name.get_attribute('innerHTML')
    review_date = comment.find_element_by_css_selector('li.upost')
    review_date = review_date.get_attribute('innerHTML')
    review_text = comment.find_element_by_css_selector('p.uopin')
    review_text = review_text.get_attribute('innerHTML')
    try:
        upvotes = comment.find_element_by_css_selector('span.uavatar-stat.upvote-count')
        upvotes = upvotes.get_attribute('innerHTML')
    except:
        upvotes = None
    rating_value = comment.find_element_by_css_selector('span.thumbs-score')
    rating_value = rating_value.get_attribute('innerHTML')
    curr_row = {
        'user_name' : user_name,
        'review_date' : review_date,
        'review_text' : review_text,
        'upvotes' : upvotes,
        'rating_value' : rating_value
    }
    df_products = df_products.append(curr_row, ignore_index=True)
df_products 
```
