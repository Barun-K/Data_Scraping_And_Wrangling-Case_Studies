# Goodreads_Website_Assignment

```python
import pandas as pd
import selenium
import requests
import re
from selenium import webdriver
```

```python
browser = webdriver.Chrome("chromedriver.exe")
browser.get('https://www.goodreads.com/choiceawards/best-books-2017')
```

```python
books = browser.find_elements_by_css_selector('div.category.clearFix')
len(books)
```

```python
df_products = pd.DataFrame(columns = ['Name of the category',
                                      'Book Name',
                                      'URL to the book cover page',
                                      'URL to the book description'])
```

```python
for book in books:
    category = book.find_element_by_css_selector('h4').get_attribute('innerHTML')
    book_name = book.find_element_by_css_selector('img.winner').get_attribute('alt')
    cover_image = book.find_element_by_css_selector('img.winner').get_attribute('src')
    desc = book.find_element_by_css_selector('a').get_attribute('href')
    
    curr_row = {'Name of the category':category,
                'Book Name':book_name,
                'URL to the book cover page':cover_image,
                'URL to the book description':desc}

    df_products = df_products.append(curr_row,ignore_index = True)

df_products
```

```python
df_book_detail = pd.DataFrame(columns = ['Author Name',
                                         'Description',
                                         'Votes',
                                         'Total Votes',
                                         'Win Percentage'])

for url in df_products['URL to the book description']:
    browser.get(url)
    
    author = browser.find_element_by_css_selector('span[itemprop="name"]').get_attribute('innerHTML')
    
    short_desc = browser.find_element_by_css_selector('div.readable.stacked.gcaBookDescription').get_attribute('innerHTML').strip()
    
    votes = browser.find_element_by_css_selector('span.greyText.gcaNumVotes').get_attribute('innerHTML')
    votes = ''.join(re.findall("\d+\,\d+", votes)).replace(',','')
    
    total_votes = browser.find_element_by_css_selector('greytext.gcaNumVotes').get_attribute('innerHTML')
    total_votes = ''.join(re.findall("\d+\,\d+", total_votes)).replace(',','')
    
    win_perc = round(int(votes)/int(total_votes)*100,2)
    
    curr_row = {'Author Name':author,
                'Description':short_desc,
                'Votes':votes,
                'Total Votes':total_votes,
                'Win Percentage':win_perc}
                                  
    df_book_detail = df_book_detail.append(curr_row,ignore_index = True)

df_book_detail    
```

```python
#merge two dataframes into one
top20_books=pd.concat([df_products,df_book_detail],ignore_index=False,axis=1)

top20_books
```

```python
top20_books.to_csv("Top 20 Books 2017.csv",index =False)
```
